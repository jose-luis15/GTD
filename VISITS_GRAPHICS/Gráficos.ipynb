### **Grafiico de indicadores de visitas**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

archivo = 'VISITAS-REPORTE-PROVINCIA.xlsx'
```

```python
data = pd.read_excel(archivo, sheet_name='PROJECTION')
data.set_index('CATEGORIAS', inplace=True)
# data = data.iloc[1:3,:]
data
```

```python
# grafico de PIE mejorado

def draw_pie(limite, mes):
    values = list(data.iloc[:,limite])
    colores = ["red", "blue"]

    explode_list = [0.15, 0] # ratios para alejar las porciones del centro.

    data.iloc[:,limite].plot(kind='pie',
                                figsize=(10, 6),      # tamaño de la figura
                                autopct=lambda p:f'{p:.1f}%, ({p*sum(values)/100 :.0f})',
                                #startangle=180,       # aungulo en el que se muestra el 1er registro (Africa)
                                colors = colores,
                                shadow=True,          # añade sombra al pie       
                                labels=None,          # no muestra etiquetas en el pie
                                pctdistance=1.25,     # ratio desde el centro hasta el inicio del texto generado por autopct 
                                explode=explode_list) # 'explode': aleja las porciones segun la lista especificada explode_list

    plt.title(f'{mes} 2023', pad=20)  # Titulo del plot
    plt.axis('equal')  # Para que el pie sea circular.
    plt.legend(labels = data.index, loc='upper left')  # añade leyenda
    plt.show()

draw_pie(0,"Marzo")
draw_pie(1,"Abril")
draw_pie(2,"Mayo")
draw_pie(3,"Junio")
```

```python
import itertools

#  Sample data
categories = data.columns.values.tolist()
bar = data.iloc[1:2,:].values.tolist()
bar_values = list(itertools.chain(*bar))
line = data.iloc[0:1,:].values.tolist()
line_values = list(itertools.chain(*line))

# Create a figure and axis object
fig, ax = plt.subplots(figsize=[8,4])

# Plot the bar chart
bar_width = 0.4
bar_positions = np.arange(len(categories)) + bar_width / 2
ax.bar(bar_positions, bar_values, bar_width, label='Ejecutadas', color = "blue")

# Add labels on top of each bar
for i, value in enumerate(bar_values):
    ax.text(i+0.2, value - 1, str(value), ha='center', fontweight='bold', color = "w")

# Plot the line above the bar chart
line_positions = bar_positions
ax.plot(line_positions, line_values, color='r', linestyle='--', label='Proyectadas') #, marker='o'

# Set labels and title
ax.set_xticks(bar_positions)
ax.set_xticklabels(categories)
# ax.set_xlabel('Periodo')
# ax.set_ylabel('Cantidad')
# ax.set_title('Cantidad de visitas proyectadas y Ejecutadas por mes')

# Add a legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
